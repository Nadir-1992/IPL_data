```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
import seaborn as sb
import pandas as pd
pd.read_csv('ipl.csv')
data_frame = pd.read_csv('ipl.csv')
sb.set()

graph = sb.scatterplot(x='T20 caps players',y='IPL',data = data_frame)
graph

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Role,IPL)", "(@Role, @IPL)"),
    ("First Name", "@First Name"),
])

# create a new plot with a title and axis labels
p = figure(title="IPL vs INTERNATIONAL", x_axis_label='Age', y_axis_label='IPL',plot_width = 700, plot_height= 700,tools=[TOOLTIPS])
p.circle('T20 caps','IPL',size = 12,source= data_frame)
show(p)
```
